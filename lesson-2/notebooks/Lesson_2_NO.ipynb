## 1. Introduksjon: Digital Etterforskning

Svaret ligger nesten alltid i **loggene**.

**Logger** er tekstfiler der datamaskiner registrerer hendelser. Hver gang du besøker et nettsted, logger inn eller får en feilmelding, skrives en tekstlinje til en fil på en server.

I dag skal vi skrive et skript for å analysere en serverlogg og finne en hacker.

## 2. Lese Filer i Python
For å analysere data må vi først få dem inn i Python. Vi bruker `open()`-funksjonen.

Vi skal lese en eksempel-loggfil som ligger i `../data/server_logs.txt`.

````python
# Definer stien til loggfilen
file_path = "../data/server_logs.txt"

# Åpne filen i 'lese'-modus ('r')
# 'with'-setningen sørger for at filen lukkes ordentlig etter at vi er ferdige
with open(file_path, "r") as file:
    # Les alle linjer inn i en liste
    logs = file.readlines()

# Skriv ut antallet leste linjer
print(f"Totalt antall linjer lest: {len(logs)}")
print("Eksempellinjer:")
# Skriv ut de første 3 linjene for å se hvordan de ser ut
for line in logs[:3]:
    print(line.strip())  # .strip() fjerner usynlige linjeskift-tegn
````

## 3. Parsing av Loggdata (String-manipulering)
En rå tekstlinje er vanskelig for en datamaskin å analysere. Vi må **parse** den—dele den opp i nyttige biter.

Se på en enkelt logglinje:
`192.168.1.10 - - [25/Mar/2026:10:00:01 +0000] "GET /index.html HTTP/1.1" 200 1024`

Den er skilt med mellomrom. Vi kan bruke `.split()` for å hakke den opp.

````python
# La oss se på den første linjen igjen
first_line = logs[0]
parts = first_line.split(" ")

print("Original:", first_line)
print("Splittede deler:", parts)

# Hente ut spesifikke data med indeks
ip_address = parts[0]
status_code = parts[8]

print(f"IP: {ip_address}, Status: {status_code}")
````

## 4. Ordbøker (Dictionaries) og Telling
Nå kan vi hente ut data, men vi vil **telle** det. Hvem besøker oss mest?

En **Dictionary** (ordbok) lagrer data i `nøkkel: verdi`-par. Det er perfekt for telling.

````python
# Enkelt ordbok-eksempel
fruktkurv = {}
fruktkurv["eple"] = 1
fruktkurv["banan"] = 5

print(fruktkurv)

# Sjekk om det finnes før vi legger til
frukt = "eple"
if frukt in fruktkurv:
    fruktkurv[frukt] = fruktkurv[frukt] + 1
else:
    fruktkurv[frukt] = 1

print("Oppdatert kurv:", fruktkurv)
````

## 5. Oppdage Angrepet (Brute Force)
Et **Brute Force-angrep** skjer når en hacker prøver mange passord for å gjette det riktige. Dette resulterer vanligvis i mange `401 Unauthorized`-statuskoder.

La oss skrive et skript for å finne IP-adressen med flest `401`-feil.

````python
mislykkede_logginn = {}

for line in logs:
    parts = line.split(" ")
    ip = parts[0]
    status = parts[8]
    
    # Vi bryr oss bare om statusen er 401 (Unauthorized)
    if status == "401":
        if ip in mislykkede_logginn:
            mislykkede_logginn[ip] += 1
        else:
            mislykkede_logginn[ip] = 1

print("Antall mislykkede innlogginger per IP:", mislykkede_logginn)
````
