## 1. Introduction: Digital Forensics
When a cyber attack happens, the first thing we ask is: *"What happened?"*
The answer is almost always in the **logs**.

**Logs** are text files where computers record events. Every time you visit a website, log in, or get an error, a line of text is written to a file on a server.

Today, we will write a script to analyze a server log and find a hacker.

## 2. Reading Files in Python
To analyze data, we first need to get it into Python. We use the `open()` function.

We will read a sample log file located at `../data/server_logs.txt`.

````python
# Define the path to the log file
file_path = "../data/server_logs.txt"

# Open the file in 'read' mode ('r')
# The 'with' statement ensures the file is closed properly after we are done
with open(file_path, "r") as file:
    # Read all lines into a list
    logs = file.readlines()

# Print the first 3 lines to see what they look like
print(f"Total lines read: {len(logs)}")
print("Sample lines:")
for line in logs[:3]:
    print(line.strip())  # .strip() removes invisible newline characters
````

## 3. Parsing Log Data (String Manipulation)
A raw line of text is hard for a computer to analyze. We need to **parse** it—break it into useful pieces.

Look at a single log line:
`192.168.1.10 - - [25/Mar/2026:10:00:01 +0000] "GET /index.html HTTP/1.1" 200 1024`

It is separated by spaces. We can use `.split()` to chop it up.

````python
# Let's look at the first line again
first_line = logs[0]
parts = first_line.split(" ")

print("Original:", first_line)
print("Split parts:", parts)

# Extracting specific data by index
ip_address = parts[0]
status_code = parts[8]

print(f"IP: {ip_address}, Status: {status_code}")
````

### Your Turn: Loop and Print
Write a loop that goes through **all** lines in `logs`. For each line:
1. Split the line into parts.
2. Print the IP address and the Status Code.

````python
for line in logs:
    parts = line.split(" ")
    ip = parts[0]
    status = parts[8]
    print(f"IP: {ip}  |  Status: {status}")
````

## 4. Dictionaries & Counting
Now we can extract data, but we want to **count** it. Who is visiting the most?

A **Dictionary** stores data in `key: value` pairs. It is perfect for counting.

````python
# Simple dictionary example
fruit_basket = {}
fruit_basket["apple"] = 1
fruit_basket["banana"] = 5

print(fruit_basket)

# Check if exists before adding
fruit = "apple"
if fruit in fruit_basket:
    fruit_basket[fruit] = fruit_basket[fruit] + 1
else:
    fruit_basket[fruit] = 1

print("Updated basket:", fruit_basket)
````

## 5. Detecting the Attack (Brute Force)
A **Brute Force Attack** happens when a hacker tries many passwords to guess the right one. This usually results in many `401 Unauthorized` status codes.

Let's write a script to find the IP address with the most `401` errors.

````python
failed_logins = {}

for line in logs:
    parts = line.split(" ")
    ip = parts[0]
    status = parts[8]
    
    # We only care if the status is 401 (Unauthorized)
    if status == "401":
        if ip in failed_logins:
            failed_logins[ip] += 1
        else:
            failed_logins[ip] = 1

print("Failed Login Counts per IP:", failed_logins)
````
